# Abstract 

Test the claim that Wines with a higher quality rating have a higher median alcohol content than wines with a lower quality rating. What are the null and alternate hypothesis 

```python
#importing libraries 
import pandas as pd 
from scipy.stats import ttest_ind
```

```python
# create a DataFrame from the data
data = []
header = ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density', 'pH', 'sulphates', 'alcohol', 'quality']

# read in the data file
with open(r'C:\Users\user\OneDrive\Desktop\Wine\winequality-red.csv') as f:
    # skip the first line since it contains the headers
    next(f)
    # loop over the remaining lines
    for line in f:
        # split the line into fields using the semicolon delimiter
        fields = line.strip().split(';')
        # convert the fields to their appropriate data types
        fixed_acidity = float(fields[0])
        volatile_acidity = float(fields[1])
        citric_acid = float(fields[2])
        residual_sugar = float(fields[3])
        chlorides = float(fields[4])
        free_sulfur_dioxide = float(fields[5])
        total_sulfur_dioxide = float(fields[6])
        density = float(fields[7])
        pH = float(fields[8])
        sulphates = float(fields[9])
        alcohol = float(fields[10])
        quality = int(fields[11])
        # create a dictionary to hold the sample data
        sample = {
            'fixed acidity': fixed_acidity,
            'volatile acidity': volatile_acidity,
            'citric acid': citric_acid,
            'residual sugar': residual_sugar,
            'chlorides': chlorides,
            'free sulfur dioxide': free_sulfur_dioxide,
            'total sulfur dioxide': total_sulfur_dioxide,
            'density': density,
            'pH': pH,
            'sulphates': sulphates,
            'alcohol': alcohol,
            'quality': quality
        }
        # add the sample to the data list
        data.append(sample)

# create a DataFrame from the data
df = pd.DataFrame(data, columns=header)
```

# Hypothesis 

* Null Hypothesis (H0): Median alcohol content of high-quality rating wines = Median alcohol content of low-quality rating wines.

* Alternate Hypothesis (H1): Median alcohol content of high-quality rating wines > Median alcohol content of low-quality rating wines.

```python
df.describe()
```

* The Maximum score of the quality of the wines in 8.
* Let low quality wines have a score less than 7
* Let High Quality wines have a score >=7

```python
# calculate the t-statistic and p-value

high_quality = df[df['quality'] >= 7]['alcohol']
low_quality = df[df['quality'] < 7]['alcohol']

t_statistic, p_value = ttest_ind(high_quality, low_quality)

print("t-statistic: ", t_statistic)
print("p-value: ", p_value)
```

# Conclusion 

Based on the results of the hypothesis test, we obtained a t-statistic of 17.82 and a very small p-value of 6.02e-65. This indicates a statistically significant difference between the median alcohol content of high-quality wines and low-quality wines. Since the p-value is much smaller than the significance level of 0.05, we can reject the null hypothesis and conclude that wines with a higher quality rating have a higher median alcohol content than wines with a lower quality rating.
